## Data Cleaning

```python
#Duplicate our original data before cleaning
data = loan_data.copy()
```

```python
#ProsperRating is the most important column in this data hence, we drop all rows with null values
data=data[data['ProsperRating (Alpha)'].notnull()]
```

```python
#Using mean to fill the missing data in the column below
data['DebtToIncomeRatio'] = data['DebtToIncomeRatio'].fillna(data['DebtToIncomeRatio'].mean())
```

```python
data.info()
```

```python
data.sample(20)
```

```python
#Renaming the column and replacing the numbers with strings
dict_list = {0 : 'Not Available', 1 : 'Debt Consolidation', 2 : 'Home Improvement', 3: 'Business', 
             4 : 'Personal Loan', 5 : 'Student Use', 6 : 'Auto', 7 : 'Other', 8 : 'Baby&Adoption', 
             9 : 'Boat', 10 : 'Cosmetic Procedure', 11 : 'Engagement Ring', 12 : 'Green Loans',
             13 : 'Household Expenses', 14 : 'Large Purchases', 15 : 'Medical/Dental', 16 : 'Motorcycle',
             17 : 'RV', 18 : 'Taxes', 19 : 'Vacation', 20 : 'Wedding Loans'}

data['ListingCategory'] = data['ListingCategory (numeric)'].map(dict_list)
data.drop(['ListingCategory (numeric)'], axis=1, inplace=True)
```

```python
data.ListingCategory.unique()
```

```python
# converting string to ordinal category type
ordinal_rating = ['AA', 'A', 'B', 'C', 'D', 'E','HR']
ordered_var = pd.api.types.CategoricalDtype(ordered = True, categories = ordinal_rating)
data['ProsperRating'] = data['ProsperRating (Alpha)'].astype(ordered_var)
```

```python
data.ProsperRating.unique()
```

```python
#Dropping ProsperRating (Alpha) column
data.drop(['ProsperRating (Alpha)'], axis=1, inplace=True)
```

```python
#Renaming the values using the str.replace method
data["LoanStatus"] = data["LoanStatus"].str.replace("FinalPaymentInProgress", "Current")
data["LoanStatus"] = data["LoanStatus"].str.replace(r"Past Due.*", "Past Due")
data["LoanStatus"] = data["LoanStatus"].str.replace("Chargedoff", "Defaulted")
#The .* is for regex expression encoding

#converting "LoanStatus" to category type
data[[ "LoanStatus"]] = data[["LoanStatus"]].astype("category")
```

```python
data.LoanStatus.dtype
```

```python
#Renaming the values using the str.replace method
data["IncomeRange"] = data["IncomeRange"].str.replace("Not employed", "$0")

# initiating the categorical class type with the pd.api method
range = ["$100,000+", "$75,000-99,999", "$50,000-74,999", "$25,000-49,999", "$1-24,999", "$0", "Not displayed"]
range_category = pd.api.types.CategoricalDtype(range, ordered= True)

#Convert datatype of IncomeRange into the initiated category above

data["IncomeRange"] = data["IncomeRange"].astype(range_category)
```

```python
#Checking IncomeRange datatype
data.IncomeRange.dtype
```

```python
# initiate the categorical class type with the pd.api method
term = [12,36,60]
term_category = pd.api.types.CategoricalDtype(term, ordered= True)

#Convert datatype of Term into the initiated category above

data["Term"] = data["Term"].astype(term_category)
```

```python
#Confirming Term datatype
data.Term.dtype
```

```python
#Converting 'EmploymentStatus' dtype to category
employ = data['EmploymentStatus'].value_counts().index
ordered_var = pd.api.types.CategoricalDtype(ordered = True,
                                           categories = employ)
data['EmploymentStatus'] = data['EmploymentStatus'].astype(ordered_var)
```

```python
#Confirming the general info of our features
data.info()
```
